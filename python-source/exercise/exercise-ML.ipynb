```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df['x'], df['y'], test_size = .2 )

X_train = X_train.astype(int)
Y_train = Y_train.astype(int)
X_test = X_test.astype(int)
Y_test = X_test.astype(int)

model = Sequential()
model.add(Dense(units = 32, kernel_initializer='normal', activation = 'relu', input_dim = 1))
#model.add(Dense(units = 64, activation = 'relu'))
#model.add(Dense(units = 1, activation = ''))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = 'accuracy')
```

```python
model.fit(X_train, Y_train, epochs = 100 , batch_size=1)
```

```python
Y_result = model.predict(X_test)
```

```python
accuracy_score(Y_test, Y_result)
```
